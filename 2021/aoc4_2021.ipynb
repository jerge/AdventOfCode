```python
x = inp.splitlines()
draw = x[0].split(",")
boards = []
for i in range(len(x[1:])//6):
  boards.append(list(map(lambda a : a.split(), x[2+i*6:1+i*6+6])))
print(draw)
boards
```

```python
def is_winning(row, drawn):
  #print(row)
  #print(drawn)
  #print((list(map(lambda x : x in drawn, row))))
  return all(list(map(lambda x : x in drawn, row)))
def has_won(board, drawn):
  for i in range(5):
    if is_winning(board[i],drawn) or is_winning([x[i] for x in board], drawn):
      return True
  #return is_winning(([board[i][i] for i in range(5)]), drawn) or is_winning(([board[-i-1][i] for i in range(5)]), drawn)
  return False
def sum_unmarked(board,drawn):
  sum = 0
  for r in board:
    for n in r:
      if not n in drawn:
        #print(n)
        sum += int(n)*int(drawn[-1])
  return sum
stop = False
for i in range(len(draw)):
  if stop == False:
    for j, board in enumerate(boards):
      if has_won(board, draw[:i]):
        stop = True
        print(f"board {j}, draw {i}")
        print(sum_unmarked(board,draw[:i]))
        break
      
```
