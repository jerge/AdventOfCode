```python
def move_dir(loc, dir):
    if dir == "L":
        return (loc[0] - 1, loc[1])
    elif dir == "R":
        return (loc[0] + 1, loc[1])
    elif dir == "U":
        return (loc[0], loc[1] + 1)
    elif dir == "D":
        return (loc[0], loc[1] - 1)
    else:
        print("Invalid dir")
    return loc
def move_tail(h,t):
    dx = t[0] - h[0]
    dy = t[1] - h[1]
    if 1 >= dx >= -1 and 1 >= dy >= -1:
        return t
    elif dx == 0 and dy > 1:
        t = move_dir(t, "D")
    elif dx == 0 and dy < -1:
        t = move_dir(t, "U")
    elif dy == 0 and dx > 1:
        t = move_dir(t, "L")
    elif dy == 0 and dx < -1:
        t = move_dir(t, "R")
    if dx >= 1 and dy >= 1:
        t = move_dir(t,"D")
        t = move_dir(t,"L")
    elif dx >= 1 and dy <= -1:
        t = move_dir(t,"U")
        t = move_dir(t,"L")
    elif dx <= -1 and dy >= 1:
        t = move_dir(t,"D")
        t = move_dir(t,"R")
    elif dx <= -1 and dy <= -1:
        t = move_dir(t,"U")
        t = move_dir(t,"R")
    return t
```
